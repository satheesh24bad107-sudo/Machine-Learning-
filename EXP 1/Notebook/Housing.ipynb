## Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Load Dataset

```python
df = pd.read_csv('Housing.csv')
df.head()
```

## Dataset Inspection

```python
df.tail()
df.info()
df.describe()
```

## Missing Values

```python
df.isnull().sum()
```

## Visualizations

```python
numeric_df = df.select_dtypes(include=['int64', 'float64'])
plt.figure()
sns.heatmap(numeric_df.corr(), annot=False)
plt.title("Correlation Heatmap of Housing Dataset")
plt.show()
```

```python
plt.figure()
plt.scatter(df['area'], df['price'])
plt.xlabel("Area (sq.ft)")
plt.ylabel("House Price")
plt.title("Scatter Plot: Area vs House Price")
plt.show()
```
