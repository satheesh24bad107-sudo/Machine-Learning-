## Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Load Dataset

```python

df = pd.read_csv('data.csv')
df.head()
```

## Dataset Inspection

```python
df.tail()
df.info()
df.describe()
```

## Missing Values

```python
df.isnull().sum()
```

## Visualizations

```python
df['Quantity'].value_counts().head(10).plot(kind='bar')
plt.show()
```

```python

df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df['Date'] = df['InvoiceDate'].dt.date
daily_sales = df.groupby('Date')['Quantity'].sum()
plt.figure()
plt.plot(daily_sales)
plt.xlabel("Date")
plt.ylabel("Total Quantity Sold")
plt.title("Daily Sales Trend Over Time")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
